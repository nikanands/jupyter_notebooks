```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data_file = pd.read_csv('~/Desktop/Dataset/Pokemon.csv')
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
data_file
```

```python
data_file.columns
```

```python
data_file = data_file.drop(['Type 2'], axis=1)
data_file
```

```python
data_file.columns = ['#', 'Name', 'Category', 'Total', 'HP', 'Attack', 'Defense',
                       'Sp. Atk', 'Sp. Def', 'Speed', 'Generation', 'Legendary']
data_file.columns
```

```python
data_file['Total']= data_file['HP']+data_file['Attack']+data_file['Defense']+data_file['Sp. Atk']+data_file['Sp. Def']+data_file['Speed']
data_file['Total']
```

```python
ordered_col_values = ['#', 'Name', 'Category', 'HP', 'Attack', 'Defense',
                       'Sp. Atk', 'Sp. Def', 'Speed', 'Total', 'Generation', 'Legendary']
```

```python
data_file = data_file[ordered_col_values]
```

```python
data_file
```

```python
data_file[data_file['Legendary']== True].index.values
```

```python
data_file['Legendary'] = np.where(data_file['Legendary'] == False, 0, 1)
pd.unique(data_file['Legendary'])
```

```python
pd.unique(data_file['Generation'])
```

```python
data_file = data_file.drop(['Generation'], axis=1)
data_file
```

```python
pd.unique(data_file['Category'])
```

```python
category_value = ['Grass', 'Fire', 'Water', 'Bug', 'Normal', 'Poison', 'Electric',
       'Ground', 'Fairy', 'Fighting', 'Psychic', 'Rock', 'Ghost', 'Ice',
       'Dragon', 'Dark', 'Steel', 'Flying']
```

```python
for i in range(len(category_value)):
    data_file['Category'] = np.where(data_file['Category'] == category_value[i],
                                    i,
                                    data_file['Category'])
```

```python
data_file
```

```python
fig_data = data_file[data_file['Category'] == 0]
pd.unique(fig_data['Total'])
```

```python
plt.figure(figsize=(16,12))
plt.scatter(x=fig_data['Name'],
           y= fig_data['Speed'],
           color='black',
           s=150)
# sns.set(rc={'axes.facecolor':'blue'})
sns.set_style('dark')
sns.despine()
plt.xticks(rotation=90)
plt.show()
```
